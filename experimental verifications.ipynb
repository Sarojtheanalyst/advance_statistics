```python
# perform the t test 
population_mean = 10
t_statistic, p_value = stats.ttest_1samp(sample, population_mean)

print("t-statistic:", t_statistic)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print(" The sample mean is significantly different from the population mean.")
else:
    print(" The sample mean is not significantly different from the population mean.")
```

```python
# to calculate the strength we have the cohens d
def cohens_d(sample, population_mean):
    sample_mean = np.mean(sample)       # Sample mean
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for sample std)
    # Cohen's d calculation
    d = (sample_mean - population_mean) / sample_std
    return d
```

```python
# modifications 
cohens_value=[]
effect=[]
for i in population_mean:
    thres=cohens_d(sample, i)
    cohens_value.append(thres)
    if thres < -1.2:  
        effect.append("Very large negative effect")
    elif -1.2 <= thres < -0.8:  
        effect.append("Large negative effect")
    elif -0.8 <= thres < -0.5:  
        effect.append("Medium negative effect")
    elif -0.5 <= thres < -0.2: 
        effect.append("Small negative effect")
    elif -0.2 <= thres < 0.2:   
        effect.append("Very small or negligible effect")
    elif 0.2 <= thres < 0.5:    
        effect.append("Small positive effect")
    elif 0.5 <= thres < 0.8:    
        effect.append("Medium positive effect")
    elif 0.8 <= thres < 1.2:    
        effect.append("Large positive effect")
    else:                   
        effect.append("Very large positive effect")
```
